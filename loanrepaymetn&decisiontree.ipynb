```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
data_set=pd.read_excel("/home/chinmaya/Desktop/coding/aiagent/ml/ml_algorithms/desisiontree&loanrepayment/loanrepayment.xlsx",header=0)
```

```python
print(data_set.head(5))
```

```python
data_set.isnull().sum()
data_set.dropna(inplace=True)
```

```python
print("data_set Lenght::"),len(data_set)
print("data_set shape::"),data_set.shape
```

```python
print(data_set.columns)
```

```python
sns.countplot(x='result', data=data_set)
plt.show()
```

```python
sns.histplot(data_set['credit_score'], kde=True)
plt.show()
```

```python
##insights
##Customers with higher credit scores are generally considered less risky, which can positively influence loan repayment behavior.
```

```python
sns.histplot(data_set['initial_payment'], kde=True)
plt.show()
```

```python
##insights
##Higher initial payments may indicate stronger financial capacity, which can reduce the likelihood of loan default.
```

```python

sns.boxplot(x='result', y='credit_score', data=data_set)
plt.title("Credit Score vs Loan Status")
plt.show()
```

```python
## Insight
##Customers who repaid loans tend to have higher credit scores compared to those who defaulted.
```

```python
## Combined Insight
##Both credit score and initial payment show diverse distributions, indicating that loan repayment behavior depends on multiple financial factors rather than a single variable.
```

```python
X= data_set.iloc[:,0:5]
y=data_set['result']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.3, random_state = 100)
```

```python
#declaring the decision tree
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,
                                   max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train,y_train)
```

```python
y_pred_en=clf_entropy.predict(X_test)
y_pred_en
```

```python
print ("Accuracy is "), accuracy_score(y_test,y_pred_en)*100
```
